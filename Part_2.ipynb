#3)Using nlp for text similarities 

```python
#Any text to find its realted entries in both datasets
ref_sent="Alisha Solid Women's Cycling Shorts"
```

```python
#Using nlp for the sentence
ref_sent_vec=nlp(ref_sent)
```

```python
#List of all product names of amazon dataset
product1=[nlp(row) for row in df1['product_name']]
```

```python
#list of all id's if amazon dataset
id1=[id for id in df1['uniq_id']]
```

```python
#creating dataset of text id and its similarity score of amazon dataset
sim_df1=[]
id_df1=[]
for i in range(len(product1)):
  sim=product1[i].similarity(ref_sent_vec)
  sim_df1.append(sim)
  id_df1.append(id1[i])
  sims_docs1=pd.DataFrame(list(zip(id_df1,sim_df1)),columns=['uniq_id','sims'])
```

```python
#Sorting the similarity percentage in descending order to find most similar product
sims_docs1=sims_docs1.sort_values(by='sims',ascending=False)
```

```python
#List of all product in flipkart dataset
product2=[nlp(row) for row in df2['product_name']]
```

```python
#List of all id's in flipkart dataset
id2=[id for id in df2['uniq_id']]
```

```python
#creating dataset of text id and its similarity score of flipkart dataset
sim_df2=[]
id_df2=[]
for i in range(len(product2)):
  sim=product2[i].similarity(ref_sent_vec)
  sim_df2.append(sim)
  id_df2.append(id2[i])
  sims_docs2=pd.DataFrame(list(zip(id_df2,sim_df2)),columns=['uniq_id','sims'])
```

```python
#Sorting the similarity percentage in descending order to find most similar product
sims_docs2=sims_docs2.sort_values(by='sims',ascending=False)
```

```python
#Creating a list of product id where have the similarity greater then 0.97 for amazon dataset
temp1=[]
for (sim, id) in zip(sims_docs1['sims'],sims_docs1['uniq_id']):
  if(sim>=0.95):
    temp1.append(id)
```

```python
#Creating a list of product id where have the similarity greater then 0.97 for flipkart dataset
temp2=[]
for (sim, id) in zip(sims_docs2['sims'],sims_docs2['uniq_id']):
  if(sim>=0.95):
    temp2.append(id)
```

```python
#If both the list have different length the choose the minimum one for both which will by then help us to create balanced dataset 
n=min(len(temp1),len(temp2))
```

```python
#Creating new excel sheet 
wb = Workbook()
ws =  wb.active
ws.title = "Changed Sheet"
wb.save(filename = 'sample_book.xlsx')
ws.append(['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart','Product name in Amazon','Retail Price in Amazon','Discounted Price in Amazon'])
```

```python
#Appending the values of product name ,discounted price and reatil price of the product from its respective amazon and flipkart datasets
j=0
for i in temp1:
  ans=[]
  if(j<=n):
    ans.append(df1.loc[df1['uniq_id']==i,'product_name'].iloc[0])
    ans.append(df1.loc[df1['uniq_id']==i,'discounted_price'].iloc[0])
    ans.append(df1.loc[df1['uniq_id']==i,'retail_price'].iloc[0])
    ans.append(df2.loc[df2['uniq_id']==i,'product_name'].iloc[0])
    ans.append(df2.loc[df2['uniq_id']==i,'discounted_price'].iloc[0])
    ans.append(df2.loc[df2['uniq_id']==i,'retail_price'].iloc[0])
    j=j+1
    ws.append(ans)
    wb.save('sample_book.xlsx')
   
  else:
    break
```
